```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn
%matplotlib inline
```

```python
import pandas as pd
import numpy as np

# Load the datasets
df1 = pd.read_csv("Year 2009-2010.csv")
df2 = pd.read_csv("Year 2010-2011.csv")

# Combine the datasets
df_combined = pd.concat([df1, df2], ignore_index=True)

# Column mapping
column_mapping = {
    "Invoice": "FOLIO_FACTURA",
    "StockCode": "IDPRODUCTO",
    "Description": "MARCA",
    "Quantity": "#UNIDADES",
    "InvoiceDate": "FECHA",
    "Price": "CPRECIO",
    "Customer ID": "IDCLIENTE",
}

# Rename columns
df_combined = df_combined.rename(columns=column_mapping)

# Keep only relevant columns
df_combined = df_combined[list(column_mapping.values())]

# Convert FECHA to datetime and extract YEAR & MES
df_combined["FECHA"] = pd.to_datetime(df_combined["FECHA"])
df_combined["YEAR"] = df_combined["FECHA"].dt.year
df_combined["MES"] = df_combined["FECHA"].dt.month  # Extracting month

# Generate synthetic data
np.random.seed(42)
df_combined["CTIPOCAM01"] = np.random.uniform(18.0, 22.0, len(df_combined))
df_combined["COSTOPESOS"] = df_combined["CPRECIO"] * df_combined["#UNIDADES"] * np.random.uniform(0.8, 1.2, len(df_combined))

# Reorder the columns to place YEAR and MES at the correct position
cols = df_combined.columns.tolist()
year_index = len(cols) - 3
mes_index = year_index + 1

cols.insert(year_index, cols.pop(cols.index("YEAR")))  # Move YEAR
cols.insert(mes_index, cols.pop(cols.index("MES")))    # Move MES right after YEAR

df_combined = df_combined[cols]

# Save to CSV
df_combined.to_csv("BASEVENTAS2010A2015.csv", index=False)

# Display first few rows
df_combined.head()
```

```python
def read_data():
    df = pd.read_csv('./BASEVENTAS2010A2015.csv', parse_dates=['FECHA'], infer_datetime_format=True)
    df.columns = df.columns.str.strip()
    df['CPRECIO'] = df['CPRECIO'].astype(str).str.replace(",", "").astype(float)
    df['COSTOPESOS'] = df['COSTOPESOS']
    df.columns.values[-4] = "YEAR"

    return df

def calculate_extra_cols(df):
    df['total_price'] =  df['CPRECIO'] * df['#UNIDADES'] * df['CTIPOCAM01']
    return df
```

```python
df = read_data()
df = calculate_extra_cols(df)
df.head()
```

#### Articles sold per year

```python
art_year = df.groupby('YEAR')[["#UNIDADES", "total_price"]].sum()
art_year['#UNIDADES'].plot(kind='bar')
```

#### Articles sold per month across all years

```python
art_month = df.groupby('MES')["#UNIDADES"].mean()
art_month.sort_index().plot(kind='bar')
```

#### Articles sold per month per year

```python
month_year = df.groupby(["YEAR", "MES"])['#UNIDADES'].sum()
month_year.plot()
# per = new.index.to_period("M")
# plt.bar(per.value_counts())
```

#### Jumping into time series

```python
ts = df.set_index("FECHA")[["#UNIDADES", "total_price"]]
ts = ts[~ts["total_price"].isnull()]
ts["total_price"] /= 100.0
ts = ts.sort_index()
axes = ts["#UNIDADES"].plot()
axes.set_xlabel("Years")
```

#### Lets try a transformation to the data

```python
logts = np.log(ts["#UNIDADES"])
axes = logts[logts > 0].plot()
axes.set_xlabel("Years")
```

#### After resampling to the sum of months 

```python
month_ts = ts.resample('M').sum().sort_index()
axes = month_ts.plot()
axes.set_xlabel("Years")
axes.legend(labels=["Units", "Price Sum / 100"])
```

```python
log_month_ts = np.log(month_ts)
axes = log_month_ts.plot()
axes.set_xlabel("Years")
axes.legend(labels=["Units", "Price Sum / 100"])
```
