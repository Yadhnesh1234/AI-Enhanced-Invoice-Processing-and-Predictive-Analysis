Feature Engineering

```python
# Convert 'date' column to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
grouped_df = df.groupby([df['InvoiceDate'].dt.date, 'StockCode'])['Quantity'].sum()
df.head()
```

```python
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
```

```python
df.head()
```

```python
from datetime import datetime
import calendar
```

```python
def weekend_or_weekday(year,month,day):
    
    d = datetime(year,month,day)
    if d.weekday()>4:
        return 1
    else:
        return 0

df['weekend'] = df.apply(lambda x:weekend_or_weekday(x['Year'], x['Month'], x['Day']), axis=1)
df.head()
```

```python
from datetime import date
import holidays

def is_holiday(x):
  
  UK_holidays = holidays.country_holidays('GB')

  if UK_holidays.get(x):
    return 1
  else:
    return 0

df['holidays'] = df['InvoiceDate'].apply(is_holiday)
df.head()
```

```python
df['m1'] = np.sin(df['Month'] * (2 * np.pi / 12))
df['m2'] = np.cos(df['Month'] * (2 * np.pi / 12))
df.head()
```

```python
from datetime import datetime

def which_day(year, month, day): 
    
    d = datetime(year,month,day) 
    return d.weekday() 

df['weekday'] = df.apply(lambda x: which_day(x['Year'], 
                                                    x['Month'], 
                                                    x['Day']), 
                                axis=1) 
df.head()
```

```python
import matplotlib.pyplot as plt

# Ensure 'Quantity' is numeric
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df = df.dropna(subset=['Quantity'])
df['Quantity'] = df['Quantity'].astype(float)

# Ensure 'StockCode' is numeric
df['StockCode'] = pd.to_numeric(df['StockCode'], errors='coerce')
df = df.dropna(subset=['StockCode'])
df['StockCode'] = df['StockCode'].astype(int)

# Ensure 'Invoice' is treated as categorical
df['Invoice'] = df['Invoice'].astype(str)

# Create 'weekend' column
df['weekend'] = df['weekday'].apply(lambda x: 1 if x >= 5 else 0)

# Features to plot
features = ['Year', 'Month', 'weekday', 'weekend', 'holidays','Day']

# Grouping & plotting
plt.subplots(figsize=(20, 10))
for i, col in enumerate(features):
    plt.subplot(2, 3, i + 1)
    df.groupby(col)['Quantity'].mean().plot.bar()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# Calculating Simple Moving Average 
# for a window period of 30 days
window_size = 30
data = df[df['Year']==2009]
windows = data['Quantity'].rolling(window_size)
sma = windows.mean()
sma = sma[window_size - 1:]

data['Quantity'].plot()
sma.plot()
plt.legend()
plt.show()
```

```python
Q1 = df['Quantity'].quantile(0.25)  # 25th percentile
Q3 = df['Quantity'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1  # Interquartile range

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df = df[(df['Quantity'] >= lower_bound) & (df['Quantity'] <= upper_bound)]

print(df.shape)  
```

```python
plt.subplot(1, 2, 2)
sb.boxplot(df['Quantity'])
plt.show()
```

```python
features = df.drop(['Quantity', 'Year','Description','Country','Customer ID'], axis=1)
target = df['Quantity'].values


X_train, X_val, Y_train, Y_val = train_test_split(features, target,
                                                  test_size = 0.05,
                                                  random_state=22)
X_train.shape, X_val.shape
```

```python
# Normalizing the features for stable and fast training.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
df.isna().sum()
```

```python
models = [LinearRegression(), XGBRegressor(), Lasso(), Ridge()]

for i in range(4):
    models[i].fit(X_train, Y_train)

    print(f'{models[i]} : ')

    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(Y_train, train_preds))

    val_preds = models[i].predict(X_val)
    print('Validation Error : ', mae(Y_val, val_preds))
    print()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, Y_train)

# Predictions
y_train_pred = rf_model.predict(X_train)
y_val_pred = rf_model.predict(X_val)

# Calculate errors
train_error_rf = mean_squared_error(Y_train, y_train_pred, squared=False)  # RMSE
val_error_rf = mean_squared_error(Y_val, y_val_pred, squared=False)  # RMSE

print("RandomForestRegressor:")
print("Training Error: ", train_error_rf)
print("Validation Error: ", val_error_rf)
```
